# Unity Catalog Object Creation
This notebook contains SQL scripts to create all required objects in the Unity Catalog


```python
%sql
CREATE CATALOG IF NOT EXISTS dev_bronze;
CREATE CATALOG IF NOT EXISTS dev_silver;
CREATE CATALOG IF NOT EXISTS dev_gold;
```

```python
%sql
-- Bronze schema
CREATE SCHEMA IF NOT EXISTS dev_bronze.r_source_h;

-- Silver schema
CREATE SCHEMA IF NOT EXISTS dev_silver.b_source_h;

-- Gold schema
CREATE SCHEMA IF NOT EXISTS dev_gold.c_data_product_h;
```

```python
%sql
CREATE VOLUME IF NOT EXISTS dev_bronze.r_source_h.source_general;
```
